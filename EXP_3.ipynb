```python
pattern = r'\s'
data['char_count']=data['twitts'].apply(lambda x: len(re.sub(pattern, '', x)))
data
```

```python
pattern = r'[!"#$%&\'()*+,-./:;<=>?@[\]^_`{|}~]'
data['twitts'] = data['twitts'].apply(lambda x: re.sub(pattern, '', x))
data
```

```python
data['hash_tags_count']= data['twitts'].apply(lambda x: len(re.findall(r'#(\w+)', x)))
data.sample(4)
```

```python
data['mentions']= data['twitts'].apply(lambda x: len(re.findall(r'#(\w+)', x)))
data.sample(10)
```
